Analisando envios

```python
#envios pedidos
df_pedidos = pd.read_csv("olist_orders_dataset.csv")
df_pedidos
```

```python
# apagando dados nulos
df_pedidos = df_pedidos.dropna()
```

```python
df_pedidos.info()
```

Verificando os produtos entregues com atraso (diferença entre data de entrega e data prevista de entrega)

```python
# transformando os dados em tipo datetime 
df_pedidos['order_estimated_delivery_date'] = pd.to_datetime(df_pedidos['order_estimated_delivery_date'], errors='ignore')
df_pedidos['order_delivered_customer_date'] = pd.to_datetime(df_pedidos['order_delivered_customer_date'], errors='ignore')

# criando novo DF para envios atrasados
colunas = df_pedidos.columns.tolist()
df_pedidos_atrasados = pd.DataFrame(columns = colunas)

# vendo os pedidos atrasados
cont_atrasados = 0
for i, row in df_pedidos.iterrows():
    # comparando as datas
    
    # se a data estimada para entrega for menor 
    # que a data de entrega, então o produto chegou com atraso 
    if(df_pedidos['order_estimated_delivery_date'][i] < df_pedidos['order_delivered_customer_date'][i]):
        df_pedidos_atrasados.loc[cont_atrasados] = df_pedidos.loc[i]
        cont_atrasados += 1
        
print(f'Quantidade de produtos com atraso na entrega: {cont_atrasados}')
```

```python
# DF de pedidos atrasados
df_pedidos_atrasados
```

Relação de entregas dentro de fora do prazo

```python
# gráfico que mostrar a relação entre produtos entregues no prazo e prdutos entregues com atraso
fig1,ax1 = plt.subplots()
valores = [len(df_pedidos_atrasados) / len(df_pedidos), (len(df_pedidos) - len(df_pedidos_atrasados)) / len(df_pedidos)]
labels = ['Entregas fora do prazo', 'Entregas no prazo']
ax1.pie(valores,labels = labels, autopct = '%1.1f%%', shadow = True, startangle = 90)
ax1.axis('equal')
plt.show()
```

Relacionando as reclamações ruins com os pedidos atrasados

```python
# unindo por order_id unicos, para saber as avaliações negativas com entrega atrasada
df_avaliacoes_ruins_entrega_atrasada = pd.merge(df_avaliacoes_ruins, df_pedidos_atrasados, how = 'inner', on = 'order_id')
df_avaliacoes_ruins_entrega_atrasada
```

```python
# gráfico que mostra a relação entre produtos com reviews negativos com entrega atrassada
# e produtos com reviews negativos que não estam com entrega atrasada
fig1,ax1 = plt.subplots()
valores = [len(df_avaliacoes_ruins_entrega_atrasada) / len(df_avaliacoes_ruins), (len(df_avaliacoes_ruins) - len(df_avaliacoes_ruins_entrega_atrasada)) / len(df_avaliacoes_ruins)]
labels = ['Avaliações ruins com entrega atrasada', 'Avaliações ruins que foram entregue no prazo']
ax1.pie(valores,labels = labels, autopct = '%1.1f%%', shadow = True, startangle = 90)
ax1.axis('equal')
plt.show()
```
